## <a name="mark_05"></a>Qué es y cómo funciona Docker?
## [Index](#index)

![](img_22.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

![](img_23.png)

- **Docker deamon o docker server**: Servicio del sistema operativo, que maneja todas las entidades con las que trabaja Docker, basicamente la que nos permite utilizar los contenedores, escalarlos, crearlos, configurarlos, distribuirlos, etc. Es quien interactua con el sistema operativo para que todo pase.

- **Rest API**: Forma en la que deamon expone una interfaz para comunicarse consigo. Nos podemos comunicar desde nuestra máquina con Docker deamon por http, o desde una máquina remota, exponiendo nuestro Docker deamon y que una máquina que este en otra red se comunique con nuestra máquina y le de instrucciones al Docker deamon (no es muy recomendable por temas de seguridad)

- **Client Docker CLI**: Cuando usamos Docker en nuestra terminal, justamente lo que estamos usando es este Cliente de Docker, siendo este Cliente quien le "habla" al Docker deamon con nuestras instrucciones. Es posible crear un cliente en python y que este cliente hable con deamon, lo mismo para otros lenguajes de programación.

- container: Donde van a correr nuestro aplicativos.

- **Image**: Artefactos que usa docker para empaquetar contenedores, y que nosotros usamos para empaquetar nuestro código, y distribuirlo a través de distintas instalaciones de Docker.

- **Data Volumes**: Forma en la que Docker nos permite acceder, con seguridad y de forma flexible, al sistema de archivos de la máquina anfitriona o servidores, sin comprometer la seguridad del mismo.

- **Network**: Le permite a los distintos contenedores comunicarse entre sí, o con el mundo exterior.

## <a name="mark_07"></a>Conceptos fundamentales de Docker: contenedores.
## [Index](#index)

Es una agrupación de procesos.

- Es una entidad lógica, no tiene el limite estricto de las máquinas virtuales, emulación del sistema operativo simulado por otra más abajo.

- Ejecuta sus procesos de forma nativa.

- Los procesos que se ejecutan adentro de los contenedores ven su universo como el contenedor lo define, no pueden ver mas allá del contenedor, a pesar de estar corriendo en una maquina más grande.

- No tienen forma de consumir más recursos que los que se les permite. Si esta restringido en memoria ram por ejemplo, es la única que pueden usar.

- A fines prácticos los podemos imaginar cómo maquinas virtuales, pero NO lo son. Máquinas virtuales livianas.

- Docker corre de forma nativa solo en Linux.

- Sector del disco: Cuando un contenedor es ejecutado, el daemon de docker le dice, a partir de acá para arriba este disco es tuyo, pero no puedes subir mas arriba.

- Docker hace que los procesos adentro de un contenedor este aislados del resto del sistema, no le permite ver más allá.

- Cada contenedor tiene un ID único, también tiene un nombre.

- Ahora corremos nuevamente "docker ps -a", podemos ver el nuevo "names" re_named

![](img_25.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Remover un contenedor en particular.
```bash
docker rm id_contenedor
```
o 
```bash
docker rm names_contenedor
```

### Remover todos los contenedores apagados.
```bash
docker container prune
```


## <a name="mark_09"></a>Corriendo Ubunto en Windows/Mac.
## [Index](#index)
```bash
docker run ubuntu
```
![](img_26.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- checkeamos el proceso.

![](img_27.png)

- "/bin/bash" --> Proceso que corre el contenedor al arrancar.
- STATUS "Exited (0)" --> Código de salida (0), el proceso terminó correctamente.
- Nos debemos conectar al shel "bash" para poder activarlo, hasta el momento el proceso "Command" no tuvo nada más que hacer, por eso se ejecuto con "/bin/bahs" y luego terminó.

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## Interactuando con Ubuntu, "it" --> Interactive.
```bash
docker run -it ubuntu
```
Nota: -i: indicamos que será interactivo -t: abre el TTY (consola)

![](img_28.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- podemos listar el sistema de archivos de Ubuntu.

![](img_29.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## Ver la distribución de linux que estamos usando:

```bash
cat /etc/lsb-release
```

![](img_30.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## <a name="mark_10"></a>Ciclo de vida de un contenedor.
## [Index](#index)


- Para el ejemplo anterior de correr ubuntu, al momento que saldo de la consola de ubuntu con "exit" el proceso "/bin/bash" se termina, ahora ¿por qué se apagó el contenedor?.
- Cada vez que un contenedor se ejecuta, se ejecuta un proceso del sistema operativo llamado "Main Proces", el que determina si el contenedor sigue activo o no. 
- Por defecto un contenedor corre siempre y cuando su Main Process este corriendo, por ese motivo cuando ejecutamos "exit" en la terminal de Ubuntu del ejemplo anterior, se corta su "Main Process" entonces el contenedor se detiene.
- Es posible dejar un contenedor sin que se apague a menos que utilize el modo interactivo, especificandole el comando con el que quiero que corra, para el siguiente ejemplo "tail -f /dev/null" (no-op)
```bash
docker run --name nombre_del_contenedor -d ubuntu tail -f /dev/null
```
- Para este ejemplo el nombre_del_contenedor es "alwaysup"
- "-d" es "--detach"
- Luego me puedo conectar a un contenedor que ya existe y que está corriendo, para ejecutar un proceso usando "docker exec" --> Ejemplo correr nuevamente la shell de ubuntu con "docker exec -it nombre_del_contenedor bash" 
```bash
docker exec -it nombre_contenedor bash
```
- Esto ejecuta nuevamente la consola de ubuntu.
![](img_31.png)

- "ps -aux" --> muestra los procesos que están corriendo en el contenedor, no todos los procesos de la máquina, por que Docker no permite que se conozcan procesos fuera del contenedor.

![](img_32.png)

- Observar que el "PID" "1" (Main Process), es "tail -f /dev/null"

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Cómo frenar "tail - /dev/null"

- Inspeccionamos el contenedor, utilizando "--format" como filtro.
```bash
docker inspect --format '{{.State.Pid}}'nombre_del_contenedor
```
- Esto devuelve un id, ejemplo "2474".

### En Linux podemos usar "kill -9 2474" para detener el contenedor.
### En Widows o Mac "docker stop nombre_contenedor" o "docker stop id_contenedor"

## <a name="mark_11"></a>Exponiendo contenedores.
## [Index](#index)

- Utilizamos el server nginx (web service y reverse proxi)
```bash
docker run -d --name proxy nginx
```

- "-d" o "--detach", el contenedor correrá en backgound, sin recibir sdt inputs o mostrar std outputs.

![](img_33.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Como vemos en la siguiente imágen el contenedor está corriendo y figura el puerto "80"

![](img_34.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Pero al intentar levantarlo, no funciona.

![](img_35.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Motivo: Cada contenedor tiene su propio "stack" de networking, su propia interfáz de red virtual, este contenedor está escuchando en el puerto "80", pero del "contenedor", que no es el puerto "80" de la máquina.
- Para este caso particulas borramos el contenedor "proxi" que creamos inicialmente y volvemos a crearlo pero con el siguiente comando:
```bash
docker run --name proxy -d -p 8080:80 nginx
```
- Donde -p es publish que conectara el puerto 80 del contenedor con el 8080 de nuestra máquina.

### Nota:  si al correr "docker run --name proxy -d -p 8080:80 nginx" no utilizamos el -d o --detach mi terminal se bloqueará mostrandome todos los std outputs del contenedor y al momento de querer detener el procesos con "ctr+c" esto también detendrá el Main Process con lo cual todo el contenedor se detendrá.

![](img_36.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Ya lo tenemos disponible en localhost:8080
![](img_37.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Si quiero ver el log del contenedor:
```bash
docker logs nombre_contenedor
```
![](img_38.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Si quiero ver los logs a medida que van llegando, uso:
```bash
docker logs -f nombre_contenedor
```
- "-f" corresponde a "follow", esto blockea mi terminal, pero estoy conectado al proceso de "logs" que no es mi proceso principal con lo cual con un "ctr+c" puedo detener ese proceso sin afectar al proceso principal del contenedor.

- Si solo quiero ver las últimas 10 líneas del log:
```bash
docker logs --tail 10 -f nombre_contenedor
```


## <a name="mark_12"></a>Bind mounts - Enlazar montajes.
## [Index](#index)

De forma predeterminada, todos los archivos creados dentro de un contenedor se almacenan en una capa de contenedor grabable. 

Esto significa que:

1. Los datos no persisten cuando ese contenedor ya no existe y puede resultar difícil sacarlos del contenedor si otro proceso los necesita.

2. La capa grabable de un contenedor está estrechamente acoplada a la máquina host donde se ejecuta el contenedor. No puede mover fácilmente los datos a otro lugar.

3. Escribir en la capa de escritura de un contenedor requiere un controlador de almacenamiento para administrar el sistema de archivos. El controlador de almacenamiento proporciona un sistema de archivos de unión, utilizando el kernel de Linux. Esta abstracción adicional reduce el rendimiento en comparación con el uso de volúmenes de datos, que escriben directamente en el sistema de archivos del host.

Docker tiene dos opciones para que los contenedores almacenen archivos en la máquina host, de modo que los archivos persistan incluso después de que el contenedor se detenga: 

- Volúmenes 
- Montajes de enlace. 

Si está ejecutando Docker en Linux, también puede usar un montaje tmpfs. 

Si está ejecutando Docker en Windows, también puede utilizar una canalización con nombre.

Documentación oficial: https://docs.docker.com/storage/

Si tenemos app que genera o usan datas y nos interesa como usuarios, poder manejar o acceder, por distitas razones:

- Compartirlos con otro contenedor
- Loging
- Persistirlos en otro proceso

### Para este ejemplo se correrá un contenedor de mongoDB, creado datos que no deben perderse al momento de terminar con el contenedor.
- Correr y crear el contenedor de mongo "name:db"
```bash
docker run -d --name db mongo
```

![](img_39.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Entrando al contenedor para generar algunos datos
```bash
docker exec -it db bash
```

![](img_40.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- El contenedor supuestamente viene con binario "mongo" que al ejecutarlo ya me conecta a la base de datos. Pero en mi caso eso no funcionó
```bash
mongo
```
- arroja:
```bash
root@d6cf75ddb9ec:/# mongo
bash: mongo: command not found
```

- Pero dentro de la carpeta "bin" se encuentra "mongosh" que también es una shell de mongo y fucniona.

**Nota**: A partir de mongo 6.0 ya no se usa "mongo" sino "mongosh"

![](img_41.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Creamos una nueva db "db_01"
```bash
use db_01
```
- Insertamos algún dato:
```bash
db.users.insert({"nombre":"leonidas"})
```
- y traemos los datos con find()
```bash
db.users.find()
```

![](img_42.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Nota: una vez que matemos el proceso de este contenedor la data no estará disponible por ese motivo, "al momento inicial de correr el contenedor" hay que "montar" la carpeta donde quiero que queden espejados los datos del contenedor, de la siguiente forma:

```bash
docker run -d --name db -v C:\Users\mende\OneDrive\Documents\Docker\docker_data:/data/db mongo
```
- Entonces estoy corriendo un contenedor que se va a llamar "db", en modo detach, con un "bind mount" que es desde "/data/db" (pertenecieste al contenedor), hacia "C:\Users\mende\OneDrive\Documents\Docker\docker_data" perteneciente al local, en mongo. 
- Para este ejemplo ya sabemos previamente que mongo guarda sus datos en "/data/db"
- "-v" hace el bind mounts.

### Ahora sí, repito los pasos anteriores para la inseción de datos

![](img_43.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- De esta forma los cambios persisten aunque elimine el contenedor, o lo detenga y luego vuelva a correrlo.

## Peligros!!! Al utilizar Bind Mounts, también le damos acceso al contenedor a la carpeta local que estamos enlazando, con lo cual hay que tener precausión en que tipo de carpeta estamos enlazando a un contenedor que no podemos validar su seguridad.

## Para evitar esto Docker utiliza los Volumenes.

## <a name="mark_13"></a>Volumenes, resolviendo problemas de seguridad.
## [Index](#index)

- La parte del disco que afecta a los volúmenes, es manejada integramente por Docker y no tenemos acceso a menos que seamos usuarios privilegiados del sistema.

- Ver los volúmenes existentes:
```bash
docker volume ls
```

![](img_44.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Podemos crea un volúmen manualmente:
```bash
docker create volume dbdata
```

![](img_45.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Ahora podemos montar el contenedor con este nuevo volumen:
```bash
docker run -d --name db --mount src=dbdata,dst=/data/db mongo
```

Donde cambiamos "-v" por "--mount", indicandole el volumen a enlazar, como source "src=dbdata", con el destino dentro del contenedor "dst=/data/db"

### En el caso de el volumen "dbdata" no exista, con la instrucción anterior Docker lo crea por nosotros, con lo cual no es necesario crearlo previo a crear el contenedor.

- Podemos confirmar este enlace inspeccionando el contenedor:
```bash
docker inspect nombre_contenedor
```
![](img_46.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Ahora ya estamos listos para trabajar con el contenedor y persistir los datos de forma segura.


## <a name="mark_14"></a>Insertar y extraer archivos de un contenedor, "bind mounts" o "volume".
## [Index](#index)

- Para este ejemplo creemos un archivo prueba.txt (vacío o no, no importa), el cual podremos mover hacia diferentes lugares del contenedor.

```bash
echo Este es un archivo de prueba > prueba.txt
```
![](img_47.png)

De esta forma el archivo ya contiene un text (cmd Windows).

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Luego activamos el contenedor con Ubunto e ingresamos en el.
```
docker start nombre_contenedor
docker exec -it nombre_contenedor bash
```
![](img_48.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Luego creamos un dir dentro del contenedor llamado "testing"

![](img_49.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Lo siguiente será introducir mi archivo "prueba.txt" dentro de "testing", algo extra, le voy a cambiar el nombre al archivo en el mismo porceso, esto es optativo.

- Para eso primero salgo de mi contenedor con "exit".

- Luego utilizo el comando "docker cp"
```bash
docker cp archivo_a_copiar contenedor_donde_sera_copiado:direccion_donde_se_copiara/nuevo_nombre
```

- Como digimos anteriormente lo migramos y le cambiamos el nombre a "test.txt"

![](img_50.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Notar que este contenedor no tiene ni bind mounts, ni volumenes, e igualmente se puede usar "docker cp"

## Ahora la extracción, tambien con "docker cp" pero del directorio "testing" dentro del contenedor.

- Parada en el directorio donde quiero hacer la extracción realizo...

```bash
docker cp nombre_contenedor:path_directorio nuevo_nombre
```
![](img_51.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Con esto vemos que no solo podemos mover o extraer archivos, o directorios.

### No hace falta que el contenedor este corriendo para usar "docker pc".

![](img_52.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Host: Maquina donde Docker esta instalado. 

- Bind Mount: Guarda los archivos en la maquina local persistiendo y visualizando estos datos (No seguro). Un ejemplo de uso es en desarrollo cuando quiero hacer cambios rápidos en mi máquina y quiero verlos reflejados en el contenedor.

- Volume: Guarda los archivos en el area de Docker donde Docker los administra (Seguro). Caso de uso, compartir datos entre contenedores, o no me quiero preocupar en que parte del disco están y quiero que persista.

- TMPFS Mount(temporary file system): Solo disponible ne Linux, es un volumen que solo existe en memoria del contenedor, es comoda si queremos utilizarla para escribir archivos temporales, que luego de apagado el contenedor serán eliminados.

## <a name="mark_15"></a>Imagenes.
## [Index](#index)

- Las imagenes son herramientas que utiliza docker para solucionar otros dos problemas, la contstrucción y la distribución de software.

- Imagenes, son plantillas o moldes a partir de las cuales Docker crea contenedores, "pieza de software empaquetada, de forma liviana que contiene todo lo necesario para que un contenedor pueda ejecutarse exitosamente, ya sea dependecia de bibliotecas nativas, código, herramientas, configuraciones,etc"

### Ver que imagenes tenemos:
```bash
docker image ls
```

![](img_53.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- El "TAG" es como la versión de la imagen, cuando no especificamos ninguna, Docker infiere "latest"

- Una imagen en el disco vive como si fuera un conjunto de archivos, o capas.

### Quiero traer una imagen a mi computadora "docker pull".

- Por defecto "docker pull" traerá las imagenes desde "docker-hub", pero se puede especificar otro repo si lo deseamos.

- Ejemplo: Buscando en Docker-Hub la imagen de Ubuntu observamos que tiene varios TAG, por cada versión.

![](img_54.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

Haremos la descarga de la misma imagen pero utilizando el TAG "22.04", analizando el listado de imagenes nos podemos dar cuenta que es exactamente la misma imagen que tiene el TAG "latest"

![](img_55.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Esto no quiere decir que tengamos "2" imagenes, lo que tenemos son 2 punteros o descripciones de la misma imagen por detrás.

- Con lo cual el espacio utilizado es solamente de "77.8MB"


## <a name="mark_16"></a>Construyendo una imagen propia.
## [Index](#index)

- El proceso de creación de imagenes siempre va a estar basado en un archivo --> "Dockerfile"

- Partiendo de un "Dockerfile", que es un archivo que describe lo que le paso al crear una imagen.

### Creando un Dockerfile

![](img_56.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Todo Dockerfile comienza con la palabra "FROM", o sea todo Dockerfile está basado en algo más, puede ser Node, Ubuntu, etc.
```
FROM ubuntu:latest
```
- Dado que partimos de una imagen de "ubuntu", versión "latest", hagamos algo más que resulte en una imagen nueva, por ejemplo correr un comando

```
FROM ubuntu:latest

RUN touch /usr/src/hola.txt
``` 

![](img_57.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- El comando RUN se ejecuta en al momento de generar o contruir la imagen, en el momento del "build", o sea que todo lo que está en "Dockerfile" se ejecuta al momento del "build".

### Construyendo la imagen con "docker build"
```bash
docker build -t ubuntu:my_version_01 .
```

- "-t" de TAG
- ubuntu, porque es una imagen de ubuntu
- my_version_01, es mi version
- "." --> finalmente le indico una ruta, "donde está el contexto de build que voy a usar", a que parte del disco tiene acceso el proceso de build mientras se está ejecutando, para este caso es el punto, o sea donde estamos parados.

**Resumen**

"Docker construime una imagen llamada ubuntu version "my_version_01" usando el directorio actual como contexto de build", y como en ese directorio donde está parado hay un Dockerfile, va a usar ese Dockerfile.

![](img_58.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Como se ve en la imagen anterior cada instrucción corrida, "FROM", luego "RUN" tiene un "id", en consecuencia para Docker, cada instrucción corrida genera una nueva "capa"/ "layer".

### Visualización de la nueva imagen creada.

![](img_59.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## Nota: Una imagen es una forma lógica de hablar de un conjunto de capas "layers".

![](img_60.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Ahora podemos correr un contenedor basado en la imagen creada.
```bash
docker run -it ubuntu:my_version_01
```

![](img_61.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Buscamos el archivo creado en el momento de la creación "docker build"

![](img_62.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Publicar imagen en Docker-Hub

- La publicación de una imagen se hace con "docker push".

- Al momento de intentar publicar esta imagen particular "ubuntu:my_version_01", esto no será posible ya que está apuntando al repositorio oficial de ubuntu, en el cual no tenemos permisos para pushear versiones oficiales.

- Con lo cual debemo "retaggear" la image a por ejemplo "leonardo/ubuntu:my_version_01", esta es la nomenclatura std en docker, para decir, quien es el dueño de esa imagen, que software contiene esa imagen y a que versión pertenece.
```bash
docker tag ubuntu:my_version_01 leomen/ubuntu:my_version_01
```
![](img_63.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Pushueando a nuestro repo personal.

- Creamos el repositorio en docker hub

![](img_64.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Nos logeamos en nuestra consola.

![](img_65.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Para pushear a nuestro repositorio la nomenclatura de nuestra imagen debe corresponderse a:
### nombre_del_repositorio/software_soporte:nombre_etiqueta

![](img_66.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## Nota: La imagen creada en el repositorio de Docke hub, obtendrá todos los datos necesarios del "software soporte", para este caso "ubuntu" (o sea esta montado sobre ubuntu), con lo cual no duplicará datos, para nuestro caso el único archivo diferente será el creado en "usr/src/" llamado "hola.txt"

## <a name="mark_17"></a>El sistema de "capas" - "layers".
## [Index](#index)

- Podemos ver las capas de las distintas imagenes de varias formas, por ejemplo directamente en docker hub

![](img_54.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

simplemente hacemos doble click sobre algun tag.

![](img_67.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- utilizando:
```bash
docker history nombre_imagen
```
### Nota: Se lee de abajo hacia arriba

![](img_68.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### "dive" como herramienta para examinar el contenedor.

- Para nuestro ejemplo si estamos usando cmd o powershell corremos...
```bash
docker run -ti --rm  -v /var/run/docker.sock:/var/run/docker.sock wagoodman/dive ubuntu:my_version_01
```
- Donde la última parte contiene el TAG de nuestra imagen.

- La imagen a continuación nos muestra los layer de la imagen, y utilizando "tab" nos movemos al panel de la izquierda.

![](img_69.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Con "ctr+u" opción de filtrado "quitar los archivos NO modificados", ve todos los archivos que fueron modificados o agregados, por "layer"

![](img_70.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## <a name="mark_18"></a>Usando Docker para desarrollar aplicaciones.
## [Index](#index)

- Primero clonamos el proyecto "https://github.com/platzi/docker"
```bash
git clone https://github.com/platzi/docker
```
- Dentro del proyecto tenemos un archivo "index.js" el cual tiene como objetivo conectarse de forma remota a una DDBB, si el resultado es satisfactorio --> "Me conecté a la DB! 😎", sino --> "💥 BOOM 💥: ' + err"

```js
const express = require('express')
const app = express()
const port = 3000

const MongoClient = require('mongodb').MongoClient

// Connection URL
const mongoUrl = process.env.MONGO_URL || 'mongodb://localhost:27017/test';

app.get('/', (req, res) => {
  MongoClient.connect(mongoUrl, { useNewUrlParser: true }, (err, db) => {
    if (err) {
      res.status(500).send('💥 BOOM 💥: ' + err);
    } else {
      res.send('Me conecté a la DB! 😎');
      db.close();
    }
  });
});

app.listen(port, () => console.log(`Server listening on port ${port}!`))
```

- Por otro lado su Docker file:

```Dockerfile
FROM node:12

COPY [".", "/usr/src/"]

WORKDIR /usr/src

RUN npm install

EXPOSE 3000

CMD ["node", "index.js"]

```
1. Nos indica que está basado en "node 12"
2. Copia todos los archivos de "este directorio" a la siguiente ruta dentro del contenedor "/usr/src/"
3. Nos paramos dentro de "/usr/src"
4. Dentro de este directorio corremos npm install (nos traemos las dependencias de node)
5. Por último exponemos un puerto (EXPOSE 3000), se puede acceder desde el local con este puerto.
6. Finalmente, definimos el comando por defecto "CMD ["node", "index.js"]" que se ejecutará al momento de "correr el contenedor" si es que no especificamos nada.

### Creando la nueva imagen.
```bash
docker build -t platziapp .
```

- Recordar que para este caso el "." indica la dirección desde donde se toma el contexto de build, o sea, "desde donde estoy parado".

![](img_71.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Corriendo contenedor.
```bash
docker run --rm -p 3000:3000 platziapp
```
- "--rm" indica --> una vez apagado el contenedor, borrarlo!

- si chequeamos en "localhost:3000" podemos ver que está respondiendo correctamente, ya que no estamos conectados a ninguna instancia de mongo.

![](img_72.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Mejorando Dockerfile performance.

```Dockerfile
FROM node:14

COPY ["package.json", "package-lock.json", "/usr/src/"]

WORKDIR /usr/src

RUN npm install

COPY [".", "usr/src/"]

EXPOSE 3000

CMD ["node", "index.js"]
```

1. Queremos crear una imagen pero construida con node14 como base.

2. Al ya haber creado una imagen anterior "node:12", Docker guarda en cache procesos que no quiero volver a repetir, por ejemplo la creación de dependcias que existen en "package.json", y "package-lock.json", pero Docker puede inferir que esos archivos ya existen y no tienen modificaciones, con lo cual no los volverá a crear.

3. Nos movemos al directorio "/usr/src" del contenedor.

4. Se corre el npm install

5. Si realizamos modificaciones en nuestro código "index.js" ahora sí queremos que se incluyan en la nueva imagen, y aunque utilicemos "." para que copie todos los archivos de esa dirección, de vuelta Docker infiere que archivos ya existen y no tienen cambios, con lo cual reutiliza lo que tiene en cache.

6. Exponemos en el puerto 3000

7. Seteamos los comando por defecto al momento de correr el contenedor.

### Segunda mejora "queremos que el contenedor se actualize cuando cambiamos nuestro código, sin rebuildear toda la imagen"

- Dentro del archivo "package.json" una de las herramientas de "nodemon" (que mide archivos de node), corre node y si alguno de sus archivos cambia vuelve a cargar esos archivos

![](img_73.png)

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Para impactar estos cambios cambiamos "CMD" en Dockerfile
```bash
FROM node:14

COPY ["package.json", "package-lock.json", "/usr/src/"]

WORKDIR /usr/src

RUN npm install

COPY [".", "usr/src/"]

EXPOSE 3000

CMD ["npx", "nodemon", "index.js"]
```

### Usando "binds mount" para impactar solo los cambios que quiero y que se actualizen automaticamente.

- Al momento de crear el contenedor hacemos el "binds mount" pero especificamente al archivo que queremos impactar, debido a que si lo hacemos con todos los archivos de esa carpeta estaremos sobre escribiendo/ renombrando los demás archivos.
```bash
docker run --rm -p 3000:3000 -v C:\Users\mende\OneDrive\Documents\Docker\docker\index.js:/usr/src/index.js platziapp
```

- De esta forma los cambios que realice en el archivo "index.js" serán impactados automaticamente en mi app, sin necesidad de rebuildear toda mi imagen.

## <a name="mark_19"></a>Docker networking: colaboración entre contenedores.
## [Index](#index)

- La forma que tiene Docker de comunicarse entre contenedores es su red virtual "docker network", algunas de ellas ya vienen por defecto.
```bash
docker network ls
```

![](img_74.png)

- "bridge" network es el default que está presente por retrocompatibilidad con versiones anteriores de Docker.

- "host" es una representación en Docker de la red real de mi máquina, es decir, si yo quiciera que un contenedor tuviera acceso a todas las interfaces de red de mi máquina, y que tenga el imp y que use los puertos de mi máquina, tendría que permitirle usar esta red.

- "none" es un tipo especial que utilizamos si queremos que un contenedor NO tenga acceso a red (desabilitar el networking).

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Creando nuestra propia red.
```bash
docker network create --attachable nombre_red
```

- "--attachable" permite conexiones entre contenedores.

![](img_75.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Inspeccionando la red.
```bash
docker network inspect nombre_red
```

![](img_76.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Conectando contenedores.
```bash
docker network connect nombre_red nombre_contenedor_a_conectar
```

![](img_78.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Como el código en "index.js" está preparado para recibir una variable de entorno, con lo cual si hay una variable de entorno llamada "MONGO_URL", tomará presedencia sobre 'mongodb://localhost:27017/test'

![](img_81.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Creamos el contenedor de la app con los parámetros de conexión.
```bash
docker run -d --name app -p 3000:3000 --env MONGO_URL=mongodb://db:27017/db_01 platziapp
```
- "--env" defino la variable de entorno al crear el contenedor "MONGO_URL".

- "mongodb://" es el protocolo de mongo.

- Luego continúa con el hostname de la máquina donde estará mongo, algo positivo es que si dos contenedores están en la misma red, pueden encontrarse por hostname usando sus nombres.

    - entonces como el nombre del contenedor de mi base de datos es "db", esto va quedando como 
        "MONGO_URL=mongodb://db"
        
    - El puerto por defecto de mongo es "27017", quedando
        "MONGO_URL=mongodb://db:27017".
        
    - Finalmente el nombre de la base de datos que quiero conectar "db_01", quedando como 
        "MONGO_URL=mongodb://db:27017/db_01"

- Finalmente "platziapp", la imagen desde la cual quiero crear un nuevo contenedor.

### Conectamos en nuevo contenedor "app" a la red "my_network_01"
```bash
docker network connect my_network_01 app
```

### Nota: Podemos comprobar las conexiones con "docker network inspect nombre_red", pero los contenedores tienen que estar activos!!!

- Como se ve a continuación, ahora sí estoy conectado a la base de datos.

![](img_82.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## <a name="mark_20"></a>Docker Compose: la herramienta todo en uno.
## [Index](#index)

- En windows y mac vienen por defecto, en Linux hay que instalarla.

- Docker compose "docker-compose.yml" nos permite describir de forma declarativo, la arquitectura de servicios que nuestra aplicación necesita, el como se tienen que comunicar entre sí, como se deben manejar los archivos, todo en un solo archivo.

![](img_83.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- "version: '3.8'", es la versión del compose file, tiene que ver con que features soporta o permite docker-compose hacer con estos servicios, es un parametro obligatorio.

- "services", indica los servicios que componen nuestra aplicación, pensando en un esquema de microservicios, o una aplicación de multples servicios. En definitiba, son las distintas partes o componentes que tiene nuestra aplicación para poder ejecutarse correctamente.

- Un servicio de compose, es parecido al concepto de un container para este caso los servicios serán "app" y "db". Entonces la diferencia con un contenedor es que un servicio puede tener uno o más contenedores de la misma imagen, con lo cual podríamos escalar la aplicación teniendo más contenedores y aumentar el ancho de banda en nuestra aplicación.

    - "app": Es el servicio que va a correr nuestra aplicación.
    
        - "image": La imagen que se utilizará para levantar el servicio.
        - "environment": Declaramos la variable de entorno "MONGO_URL".
        - "depend_on": El contenedro dependerá del servicio "db", si "db" no se puede levantar, entonces "app" tampoco, y que "db" necesita ser levantado antes que "app"
        - "port": exponer el puerto 3000 del contenedor al puerto 3000 de nuestra máquina anfitriona.
    
    - "db": Es el servcio que va a correr nuestra base de datos.
    
### Ejecutando docker-compose:

- parado en la dirección donde está mi "docker-compose.yml" correr:
```bash
docker-compose up
```

- Como se ve a continuación, se crea:

    - La "network_default"
    - El contenedor para la base de datos "docker-db-1"
    - El contenedor para la aplicación "docker-app-1"
    - Se conectan los dos contenedores "Attaching to docker-app-1, docker-db-1"

![](img_83.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- si no quiero ver todos los logs puedo correr con "-d"
```bash
docker-compose up -d
```
### Nota: En este ejemplo el contenedor creado "docker-db-1" muestra en "port" 27017/tcp, esto indica que el puerto expuesto del contenedor es el 27017, pero no me puede conectar desde mi local a menos que ingrese a travez de la red de Docker. Este puerto está expuesto solo para que se comuniquen los contenedores entre sí

## <a name="mark_21"></a>Subcomandos de Docker Compose.
## [Index](#index)

* **"docker network ls"** (listo las redes) 

* **"docker network inspect docker_default"** (veo la definición de la red) 

* **"docker-compose logs"** (veo todos los logs)

* **"docker-compose logs app"** (solo veo el log de "app")

* **"docker-compose logs -f app"** (hago un follow del log de app)

* **"docker-compose exec app bash"** (entro al shell del contenedor app)

* **"docker-compose ps"** (veo los contenedores generados por docker compose)

* **"docker-compose down"** (borro todo lo generado por docker compose)


## <a name="mark_22"></a>Docker Compose como herramienta de desarrollo.
## [Index](#index)

- Para nuestro ejemplo nosotros estamos usando la imagen "platziapp" creada anteriormente, la cual es inmutable.

![](img_83.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### Necesito escribir código o modificar y que esos cambios pasen automaticamente a mi contenedor, sin hacer un build cada vez que realizamos cambios.

- De forma manual si deseo que los cambios en mi local se reflejen en el contenedor primero tendría que construir nuevamente la imagen con "docker-compose build up" o "docker-compose build up app" (indicandole solo el servicio que quiero rebuildear), y luego cargar los cambios en mi contenedor con "docker-compose up -d".

- Para evitar esto vamos a usar el bind mount dentro del "docker-compose.yaml"

![](img_85.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Pero podemos colocar el path donde se encuentre nuestro contexto, reemplazando "image" por "build": "path", y si nuestro contexto está en la misma carpeta podemos usar ".""

- "MONGO_URL" en el 27017 port, y conectando a la "db_01"

- volumes:
      - .:/usr/src (copiamos lo que tenemos en /usr/src dentro del contenedor)
      - /usr/src/node_modules (En la copia, ignoramos node_modules folder)
      
- command: npx nodemon index.js (es el comando que monitorea los cambios en "index.js")

- Ahora a buildear nuestra imagen con:
```bash
docker-compose build 
```

- Esto crea la imagen, y construye su nombre de la siguiente forma:

    - nombre_carpeta_del_contexto:latest


## <a name="mark_23"></a>Compose en equipo: override.
## [Index](#index)

- docker-compose.override.yml es un archivo que se encarga de sobreescribir tu configuración de docker-compose.yml, debido a que durante nuestro trabajo podemos necesitar una modificando continua de nuestro ambiente, y esto, si estamos en un equipo significaría que los cambios de nuestras pruebas no se pusheen a la rama Master y que el docker-compose.yaml se mantenga como fue definido para el equipo.

- "docker-compose.override.yml", es simplemente un compose file, que sirve para hacer cambios que impactan sobre "docker-compose.yml"

- Creamos el archivo "docker-compose.override.yml", usando este nombre docker lo detecta automaticamente, podríamos cambiar el nombre pero hay que decirle a docker como tiene que encontrarlo

- Pueden tener diferentes overrides por enetorno por ejemplo:

**docker-compose.override.yml para development**

**docker-compose.staging.yml para staging**

**docker-compose.production.yml para production**

Y debes correrlos de la siguiente forma:

**para development:**
```bash
docker-compose up
```
**para staging:**
```bash
docker-compose \
	-f docker-compose.ym \
	-f docker-compose.staging.yml \
	up
```
**para production:**
```bash
docker-compose \
	-f docker-compose.yml \
	-f docker-compose.production.yml \
	up
```

### Hay que tener en cuenta que "docker-compose.override.yml" se combina con su padre, "docker-compose.yml" en algunos casos sobre escribiendo la información del padre y en otras mergeandola.

![](img_86.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

![](img_87.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- En el ejemplo anterior al momento de correr "docker-compose build" app tiene que ser buildeada y taggeandola como "platziapp:latest", debido a que el docker compose padre indica image: platziapp.

### Notar que al momento del build docker usa los parametros originales del padre (que no estan sobre escritos).

- Otro ejemplo sería si quiero generar otra env variable diferente
```bash
services:
  app:
    build: .
    environment:
      MONGO_URL_02: "mongodb://db:27017/db_02"
```
### Esto genera una variable extra sin sobre escribir la origina, si quiero sobre escribirla tengo que usar el mismo nombre, ejemplo MONGO_URL

### No es aconsejable usar los "ports" en 2 compose files diferentes, docke compose no puede resolver eso, entonces definirlo solo en uno (el padre).

### Una buena práctica sería agregarlo al "gitignore" para que no sea comiteado por accidente.

## <a name="mark_24"></a>Escalando contenedores.
## [Index](#index)

- Previamente tengo que dejar explicito en mi "docker-compose.yml" los puetos que estarán dispoinibles para los contenedores.
```yml
version: "3.8"

services:
  app:
    build: .
    environment:
      MONGO_URL: "mongodb://db:27017/db_01"
    depends_on:
      - db
    ports:
      - "3000-3005:3000"
    volumes:
      - .:/usr/src
      - /usr/src/node_modules
    command: npx nodemon --legacy-watch index.js

  db:
    image: mongo
```

- Con este cambio dejo disponible desde el puerto 3000 al 3005 para escalarlo, y al momento de crear los contenedores los escalo.
```bash
docker-compose up -d --scale app=5
```

Recordar que "app" es el nombre de la aplicación a escalar.

## <a name="mark_25"></a>Administrando tu ambiente de Docker.
## [Index](#index)

- Eliminar los contenedores apagados "docker container prune"

- Eliminar contenedores apagados y encendidos "docker rm -f $(docker ps -aq)", donde el parámetro q es el id del contenedor, "docker ps -a -q" itera sobre todos los contenedore trayendo su id.

- "docker network prune" borra las network que no se estan usando.

- "docker volume prune" borra los volumenes que no estan en uso.

- "docker system prune" borra en conjunto containers, network, volumenes e imagenes (dangling, las que no están tagueadas).

### Limitando los recursos de los contenedores a 1GB, ejemplo.

```bash
docker run -d --name app --memory 1g platziapp
```

- "docker stats" visualización de cuantos recursos se utilizan por nuestros containers.

![](img_88.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Si el contenedor no tiene suficiente memoria, realizando un "docker inspect nombre_contenedor", podemos ver como en este ejemplo que esta OOMKilled, muerto por Out Of Memory.

![](img_89.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## <a name="mark_26"></a>Deteniendo contenedores correctamente: SHELL vs. EXEC.
## [Index](#index)

- Contexto de la lección modo SHELL vs. modo EXEC 

- Al estar activo o recién crear un contendor, este debería estar ejecutando un proceso principal o main process para mantenerse en funcionamiento. En caso de que el main process se detenga el contendor debería dejar de funcionar.

- Docker tiene una manera de manejar los procesos de los contenedores de manera estándar, cuando se manda detener el contenedor con la instrucción "docker stop name-container"; con esta instrucción Docker manda una señal estándar de linux llamada SIGTERM al proceso; después de un tiempo el proceso deberá detenerse; en caso de no detenerse el proceso enviará otra señal llamada SIGKILL; con esta señal se garantiza que el proceso se ha terminado de manera forzada.

- Existen problemas en la terminación de procesos ejecutados en los contenedores, una de las causas está asociado a ¿Cómo se debería declarar en Dockerfile el proceso a ejecutar?

digamos que tenemos estos dos archivos configurados de la siguiente manera

- El primero es el dockerfile; sirve para construir nuestra imagen y
```Dockerfile
FROM ubuntu:trusty
COPY ["loop.sh", "/"]
CMD /loop.sh
```
- El segundo es el archivo bash (loop.sh) que se ejecutará como un main process o proceso principal del contenedor. 
```Bash
#!/usr/bin/env bash
trap 'exit 0' SIGTERM
while true; do :; done
```

- Es importante destacar que la línea de CMD ejecuta nuestro archivo bash, a través de un shell, esta forma de expresar esta línea se conoce como shell form y esto genera que al momento de detener un contenedor que esté ejecutando un proceso por shell form necesariamente tendrá que usar un SIGKILL. 

- Para evitar esto, el CMD debe ejecutar nuestro main process de form exec form , es decir nuestro archivo debe estar configurado de esta manera.
```Dockerfile
FROM ubuntu:trusty
COPY ["loop.sh", "/"]
CMD ["/loop.sh"]
```

- La idea de que el archivo esté configurado de esta manera es que de tiempo a que el contenedor procese las solicitudes restantes y pueda hacer un Great full shutdown. 

- Otra nota importante es que los códigos de salida mayor a 128 es el resultado de una salida por un código de excepción.

* comando utilizados en clases++

**cd ~/docker/avanzado**

**docker build -t loop . build -t nombre_imagen contexto_de_construccion** --> Construir la imagen se le coloca un tag de loop

**docker run -d --name looper loop run -d --name nombre_contenedor nombre_imagen** --> Crear un contenedor con un nombre basado en una imagen

**docker stop looper stop nombre_contenedor** --> Envía la señal SIGTERM al contenedor

**docker ps -l** --> Muestra el último proceso ejecutado

**docker kill looper Kill nombre_contenedor** --> Envía la señal SIGKILL al contenedor

**docker exec looper ps -ef exec nombre_contenedor ps -ef** --> Muestra los procesos del contenedor

## <a name="mark_27"></a>Contenedores ejecutables: ENTRYPOINT vs CMD.
## [Index](#index)

- A veces queremos utilizar un contenedor como un ejecutable, solamente para correr una pequeña tarea sin tener que estar instalando los binarios en nuestro local.

- Para nuestro ejemplo utilizamos la carpeta "ping"

![](img_90.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Donde este Dockerfile, llama al binario "ping" (manda un paquete a otra máquina), hace 3 pings (con "-c", que es count y "3") a "localhost", en este caso localhost es el contenedor.

- Explicado lo anterior, construimos la imagen con "docker build -t ping ."

- Seguido corremos el contenedor con "docker run --name pinger ping", podemos ver las 3 llamadas al contenedor127.0.0.1

![](img_91.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Entonces ¿Qué pasa si quicieramos usar este contenedor como un binario de ping que puedo usar en cualquier máquina?, queriamos, pasarle parámetros para hacer ping sobre eso, ejemplo "docker run --name pinger ping hostname", con esto estoy pegandole a mi máquina.

- Como podemos ver en la siguiente imagen, al correr el comando sigue pensando que el hostname es el contenedor, entonces hay que modificar.

- Otra cosa es que al definer el command "hostname" cuando corro el contenedor, estoy sobre escribiendo el CMD del Dockerfile.

![](img_92.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

- Entonces, ¿Cómo hacemos para tener un contenedor que sea ping, pero que acepte parametros por linea de comando y si no recibe nada, haga algo?

### Utilizando ENTRYPOINT

- El ENTRYPOINT de un contenedor es el comando por defecto que se va a correr siempre, a menos que lo especifiquemos de otra manera, utilizando lo que diga CMD, como parámetro del ENTRYPOINT.

- Hago nuevamente un build "docker build -t ping ."

- Corro nuevamente un contenedor (borrarlo antes) con un nuevo parámetro "google.com"

![](img_93.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## <a name="mark_28"></a>El contexto de build.
## [Index](#index)

- Cada vez que vamos a hacer un build de una imagen, Docker va a montar en un file system temporal, todos los archivos que halla disponible, en la ruta que le paso como último parámetro a Docker build, hasta el momento le pusimos "." porque es el directorio donde estoy parado. Pero lo que puede suceder es que como estamos copiando TODO, puede haber cosas que no quiero copiar.

- Bueno, finalmente es usar ".dockerignore"

## <a name="mark_29"></a>Multi-stage build.
## [Index](#index)

Documentación https://docs.docker.com/build/building/multi-stage/

- No quiero que todo el código se encuentre en la imagen productiva final.

    - Si usamos lenguajes compilandos donde el código se convierte a binarios, o sea a un artefacto ejecutable, no quiero que todo ese código finalmente termine en mi imagen productiva, solo quiero que el artefacto ejecutable se quede ahí.
    - Otro caso tipico es el código de test, quiero correr el testing con cada build, pero no quiero que esté en la imagen productiva.
    
- Vamos a la carpeta build de nuestro proyecto.

![](img_94.png)
**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------**

### development.Dockerfile

```Dockerfile
FROM node:12

COPY ["package.json", "package-lock.json", "/usr/src/"]

WORKDIR /usr/src

RUN npm install

COPY [".", "/usr/src/"]

EXPOSE 3000

CMD ["npx", "nodemon", "index.js"]
```

### poduction.Dockerfile
```Dockerfile
FROM node:12 as builder

COPY ["package.json", "package-lock.json", "/usr/src/"]

WORKDIR /usr/src

RUN npm install --only=production

COPY [".", "/usr/src/"]

RUN npm install --only=development

RUN npm run test


# Productive image
FROM node:12

COPY ["package.json", "package-lock.json", "/usr/src/"]

WORKDIR /usr/src

RUN npm install --only=production

COPY --from=builder ["/usr/src/index.js", "/usr/src/"]

EXPOSE 3000

CMD ["node", "index.js"]
```

- El Dockerfile de producción contiene 2 “fases de build” que se pueden pensar como hacer 2 build seguidos, en donde al final la imagen construida contendrá lo especificado en el ultimo de los build.

- El primer build corre 1 test que verifica que todo funcione bien El segundo build construye la imagen final aprovechando el caché de las capas del primer build.

- Al final el 2do build es solo una extracción de lo que nos interza del primer build.

- Lo importante en este caso especifico es que si el test falla, entonces el build 2 no se corre, lo que significa que la imagen no se construye.

### Creando la imagen "docker build -t prodapp - build/production.Dockerfile ."

### Creando contenedor "docker run -d --name prod prodapp"

- Podríamos tomar el script de "production.Dockerfile", como un script de CI, Integración Continua, porque si cualquiera de los steps falla, se va a detener el build tirando un error.

## <a name="mark_30"></a>Docker-in-Docker
## [Index](#index)

- Utilizando docker desde un contenedor, o sea que desde una misma app se pueda correr scripts que levanten contenedores.

- Existe la Posibilidad de usar Docker desde otros contenedores, se logra usando el Docker socket con bind mount se accede a el archivo docker sock a la maquina anfitriona y accediendo a el desde el otro Docker el cliente puede accederlo puede hablarle directamente.

- Para Linux y Mac
```bash
docker run -it --rm -v /var/run/docker.sock:/var/run/docker.sock docker:19.03.12 
```

-Para windows
```bash
docker run -it --rm -v //var/run/docker.sock:/var/run/docker.sock docker:latest
```

### Utilización practica con "dive".

- Entonces montando el docker socket a un contenedor y entender a  un contenedor como algo ejecutable "siendo un contenedor que explora el estado de docker"

```bash
docker run --rm -it -v /var/run/docker.sock:/var/run/docker.sock -v $(which docker):/bin/docker wagoodman/dive:latest prodapp
```

### Resumen de comandos.
**docker ps** --> muestra los contenedores que están corriendo.

**docker ps -a** --> muestra todos los contenedores y sus estados.

**docker run nombre_contenedor** --> corre y crea contenedores.

**"docker inspect id_contenedor" o "docker inspect el_names_del_contenedor"** --> ver fallos o configuración.
**docker run --name new_nick_name name_contenedor_existente** --> nuestro propio nick name.

**docker rename new_nick_name re_named** --> renombrando contenedor.

**docker rm id_contenedor** --> removiendo contenedor.

**docker container prune** --> removiendo contenedores apagados.

**docker run -it ubuntu** --> interactuando con ubuntu.

**ls -lac** --> listar ordenadamente.

**docker run --name nombre_del_proceso -d ubuntu tail -f /dev/null** --> deja un contenedor corriendo sin apagarse hasta que indiquemos lo contrario.

**docker exec -it nombre_contenedor bash** --> correr un proceso en un contenedor que ya existe y que está corriendo.

**docker inspect --format '{{.State.Pid}}'nombre_del_contenedor** --> Inspeccionamos el contenedor, utilizando "--format" como filtro.

**kill -9 numero_proceso** --> Matar un proceso en Linux.

**docker stop nombre_o_id_contenedor** --> Detiene el contenedor.

**docker run --name proxy -d -p 8080:80 nginx** --> crear y publicar el puerto 80 del contenedor en el 8080 de nuestro local.

**docker logs nombre_contenedor** --> Ver logs.

**docker logs -f nombre_contenedor** --> Ver logs a medida que llegan.

**docker logs --tail 10 -f nombre_contenedor** --> Ver las últimas 10 líneas del log.
